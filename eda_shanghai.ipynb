```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat


gt_path = "ShanghaiTech/part_B/train_data/ground-truth"

counts = []

for file in os.listdir(gt_path):
    if file.endswith(".mat"):
        mat = loadmat(os.path.join(gt_path, file))
        
        # Lấy số lượng head annotations
        points = mat["image_info"][0][0][0][0][0]
        count = len(points)
        
        counts.append(count)

counts = np.array(counts)

#Thống kê 
min_count = np.min(counts)
max_count = np.max(counts)
mean_count = np.mean(counts)

print("Min:", min_count)
print("Max:", max_count)
print("Mean:", mean_count)

#Vẽ Histogram
plt.hist(counts, bins=30)
plt.title("Histogram of People Counts")
plt.xlabel("Number of People")
plt.ylabel("Number of Images")
plt.savefig("histogram.png")
plt.show()

#Lưu
with open("eda_summary.txt", "w") as f:
    f.write(f"Total images: {len(counts)}\n")
    f.write(f"Min: {min_count}\n")
    f.write(f"Max: {max_count}\n")
    f.write(f"Mean: {mean_count:.2f}\n")
    
```
